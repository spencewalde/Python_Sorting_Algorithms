```python
######## Ascending order ##########:
def bubblesortAsc(arr):
    
    n = len(arr)
    temp = 0
  
    for j in range(n-1,0,-1):
        for i in range(j):
            if arr[i] > arr[i+1]:
                temp =arr[i]
                arr[i]=arr[i+1]
                arr[i+1]=temp



    return arr
```

```python
######## Descending order #######
def bubblesortDec(arr):

    n = len(arr)
    temp = 0
  
    for j in range(n-1,0,-1):
        for i in range(j):
            if arr[i] < arr[i+1]:
                temp =arr[i]
                arr[i]=arr[i+1]
                arr[i+1]=temp


    return arr
```

```python
######## insertion sort ########
def insertionsort(arr):

    n = range(1,len(arr))

    for i in n:
        vsort = arr[i]
        for j in range(i-1,-1,-1):
            if(vsort <= arr[j] and j == 0):
                arr[j+1] = arr[j]
                arr[j] = vsort
                break
            if(arr[j] > vsort):
                arr[j+1] = arr[j]
            else:
                arr[j+1] = vsort
                break
    


    return arr
```

```python
#####selection sort #########
def selectionsort(arr):
    n = range(0,len(arr) -1)
    for i in n:
        minIndex = i

        for j in range(i+1,len(arr)):
            if arr[j] < arr[minIndex]:
                minIndex = j
        if minIndex != i:
            arr[i], arr[minIndex] = arr[minIndex],arr[i]

            
    return arr
```

```python
##### Merge Sort ####
## uses recursion better for large datasets
def mergesort(arr):

    mergesort2(arr, 0, len(arr)-1)

    return arr

def mergesort2(arr, first, last):
    if first < last:
        middle = (first + last)//2
        mergesort2(arr, first, middle)
        mergesort2(arr, middle+1, last)
        merge(arr, first, middle+1, last)

def merge(arr, first, middle, last):
    left = arr[first:middle]
    right = arr[middle:last+1]

    left.append(999999)
    right.append(999999)

    ## i for left :: j for right ##
    i = j = 0
    for k in range(first,last+1):
        if left[i] <= right[j]:
            arr[k] = left[i]
            i += 1
        else:
            arr[k] = right[j]
            j += 1
  
```
